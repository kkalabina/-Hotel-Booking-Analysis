## **Аналіз бронювання готелів**

### Опис набору даних

Набір даних Hotel Bookings – це повний набір інформації про бронювання готелів, скасування бронювання та відомості про гостей. Цей набір даних дає уявлення про різні аспекти, такі як тип готелю, кількість дорослих, дітей і немовлят на бронювання, тривалість перебування у вихідні (субота або неділя) і будні (з понеділка по п’ятницю), вибраний план харчування. за гостями, країною їх походження та позначенням сегмента ринку. Крім того, цей набір даних містить значну інформацію про статус бронювання та його оновлення. Він охоплює інформацію про те, чи було скасовано бронювання чи ні, проміжок часу між датою бронювання та датою прибуття, номер тижня та день прибуття. Він також вказує, чи є гість повторним відвідувачем чи новим клієнтом. Набір даних також містить інформацію, пов’язану з розподілом кімнат. У ньому згадуються як зарезервовані типи кімнат (тип кімнати, який спочатку запитувався), так і призначені типи кімнат (фактично виділена кімната). Крім того, він розкриває будь-які зміни, внесені в бронювання, а також подробиці про попередні скасування, зроблені гостями. Іншими важливими факторами в цьому наборі даних є тип депозиту для кожного бронювання; Ідентифікаційні номери туристичних агентств, які використовуються для бронювання; дні, проведені в списках очікування до підтвердження; класифікація споживачів, наприклад тимчасова або групова; середня добова ставка, розрахована на основі операцій з розміщенням, поділених на загальну кількість ночей; необхідні місця для паркування автомобілів, вказані замовниками; загальна кількість особливих запитів, зроблених кожним гостем.

### Профіль колонок набору даних:

- **index**
- **hotel**: тип готелю (курортний або міський)
- **is_canceled**: чи було бронювання скасовано чи ні (0 не скасовано, 1 скасовано)

- **arrival_date_month**: місяць прибуття

- **stays_in_week_nights**: показує, скільки днів тижня (з понеділка по п’ятницю) гості зупинялися або забронювали проживання в готелі протягом робочих днів

- **children**: кількість дітей, включених у кожне бронювання
- **babies**: кількість немовлят, включених у кожне бронювання

- **country**: позначає країну походження кожного гостя, який зробив бронювання

- **distribution_channel**: указує різні канали, за допомогою яких було здійснено бронювання (наприклад, туристичні онлайн-агентства, пряме бронювання в готелях/туроператорах, корпоративні домовленості)
- **is_repeated_gues**: вказує, чи є гість повторним відвідувачем (0 неповторний гість, 1 повторний гість)
- **previous_cancellations**: кількість разів, коли гості раніше скасовували свої бронювання
- **previous_bookings_not_canceled**: кількість попередніх бронювань, зроблених гостями, які не були скасовані
- **reserved_room_type**: тип спочатку зарезервованої кімнати
- **assigned_room_type**: тип кімнати, призначеної для гостей
- **booking_changes**: кількість змін, внесених до бронювання
- **deposit_type**: вказує на тип депозиту, внесеного для бронювання
- **agent**: ідентифікатор туристичної агенції, яка здійснила бронювання
- **company**: ідентифікатор компанії, яка здійснила бронювання
- **days_in_waiting_list**: кількість днів, протягом яких бронювання було в списку очікування до підтвердження
- **customer_type**: тип клієнта (наприклад, тимчасовий, договірний, груповий або інший)
- **adr**: середня денна ставка (ціна за номер) для бронювання

- **reservation_status**: вказує статус бронювання (наприклад, скасовано, зареєструвалося, не прибуло)
- **reservation_status_date**: позначає дату останнього оновлення статусу бронювання

#### 3. Питання 1: Які типи готелів (курортний або міський) мають найвищий рівень попиту? 

```python
# Підрахунок кількості бронювань для кожного типу готелю
demand_by_hotel = df['hotel'].value_counts().reset_index()
demand_by_hotel.columns = ['hotel', 'num_bookings']

# Побудова графіку
sns.barplot(data=demand_by_hotel, x='hotel', y='num_bookings', hue='hotel')
plt.title("Рівень попиту на курортні та міські готелі")
plt.xlabel("Тип готелю")
plt.ylabel("Кількість бронювань")
plt.show();
```

**Висновки:** Міські готелі користуються більшим попитом, оскількість бронювань для міських готелів значно перевищує попит на курортні готелі. Це може бути зумовлено тим, що міські готелі частіше обирають для бізнес-подорожей, коротких поїздок чи відряджень, тоді як курортні готелі більш популярні серед туристів у період відпусток.

Можливі причини різниці в попиті:
Міські готелі зручно розташовані в містах, де більший попит через бізнес, конференції чи короткі подорожі. У них менший вплив сезонності, оскільки використовуються для робочих і ділових цілей протягом усього року. Курортні готелі навпаки популярні у високі сезони (літо, зима), але попит значно падає у міжсезоння. Також курортні бронювання зазвичай довші, але кількість клієнтів може бути нижчою.

Для курортних готелів можна розробити програми, щоб стимулювати попит у міжсезоння (наприклад, знижки, спеціальні пропозиції для сімей або корпоративних клієнтів). Також можна зменшити частку скасованих бронювань через обмеження на безкоштовне скасування або введення депозитів. А для міських готелів продовжувати залучати ділових мандрівників через партнерства з компаніями та бізнес-пропозиції. Також можна забезпечити максимальну гнучкість для коротких бронювань, що важливо для клієнтів міських готелів.

Ще варто проаналізувати такий самий графік, але виключити скасовані бронювання, ля того щоб оцінити реальний попит.

```python
# Фільтрація: виключення скасованих бронювань
confirmed_bookings = df[df['is_canceled'] == 0].copy()

# Підрахунок кількості нескасованих бронювань для кожного типу готелю
demand_by_hotel_1 = confirmed_bookings['hotel'].value_counts().reset_index()
demand_by_hotel_1.columns = ['hotel', 'num_bookings']

# Побудова графіка
sns.barplot(data=demand_by_hotel_1, x='hotel', y='num_bookings', hue='hotel')
plt.title("Рівень попиту на курортні та міські готелі (без скасованих бронювань)")
plt.xlabel("Тип готелю")
plt.ylabel("Кількість нескасованих бронювань")
plt.show();
```

**Висновки:** Міські готелі мають значно вищий реальний попит порівняно з курортними готелями. Однак для обох типів готелів є можливість підвищення прибутковості через зменшення скасувань і розвиток маркетингових стратегій для низького сезону. Для більш детального аналізу можна проаналізувати попит із розподілом по місяцях або сезонах, щоб визначити пікові та слабкі періоди для кожного типу готелів.

#### 4. Питання 2: Аналіз середньої денної ставки (ADR)

```python
# Зведена таблиця ADR за типом клієнтів і каналом бронювання
adr_analysis = pd.pivot_table(
    df, 
    values='adr', 
    index=['hotel', 
           'customer_type'], 
    columns='distribution_channel', 
    aggfunc='mean'
).round(2)

# Виведення результату
adr_analysis
```

```python
# Значення NaN заважають сприймати отриману таблицю, тому замінимо їх на 0
adr_analysis.fillna(0)
```

**Висновки:** Transient клієнти приносять найбільший дохід, особливо через канал Direct. Це підтверджує важливість залучення тимчасових клієнтів. Міські готелі мають вищий ADR, ніж курортні готелі. Прямі канали бронювання (Direct) — найбільш прибуткові для всіх типів клієнтів. Контрактні клієнти та корпоративні бронювання мають найнижчий ADR. Це можна очікувати через знижки та домовленості. Там, де ми отримали NaN, або після заміни 0, для легшого сприйняття, виходить що такої комбінації серед даних не існує.

Для покращення ситуації можна зосередитись на залученні Transient клієнтів через прямі канали. Це найбільш прибутковий сегмент. Оптимізувати маркетинг для міських готелів, де ADR вище. Переглянути умови співпраці з корпоративними клієнтами, щоб підвищити прибутковість. Розробити стратегії для підвищення ADR у курортних готелях, наприклад, через додаткові послуги або преміальні пакети.

#### 5. Питання 3: Перевірка бронювань на сезоність 

```python
# Створення нової колонки з датою
df['date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' + df['arrival_date_month'], format='%Y-%B')

# Групування за роком і місяцем та підрахунок бронювань
monthly_bookings = df.groupby('date').size()

# Побудова графіку
plt.figure(figsize=(12, 6))
plt.plot(monthly_bookings.index, monthly_bookings.values)
plt.xlabel('Дата')
plt.ylabel('Кількість бронювань')
plt.title('Сезонність бронювань за роками')
plt.grid(True)
plt.xticks(rotation=45)  # Повертаємо мітки по осі X для кращої читабельності
plt.show();
```

**Висновки:** Найбільша кількість бронювань спостерігається в літні місяці (липень-серпень), що можна пояснити високим туристичним сезоном. Найменша кількість бронювань припадає на початок року (січень-лютий), що збігається із зимовим міжсезонням. 

Загальна тенденція демонструє, що з часом (з 2015 по 2017 рік) кількість бронювань зростає. Це може свідчити про популяризацію послуг готелів, зростання туризму або розширення бази даних готелів.

Готелі можуть підлаштовувати кількість персоналу та рекламу відповідно до сезонних коливань. Наприклад, збільшити рекламну активність напередодні літа. 

У міжсезоння (січень-лютий та вересень-жовтень) доцільно пропонувати акції та знижки для залучення клієнтів. Якби була інформація про розташування готелів, можна було би більш детально оцінити, чому в ці періоди падає кількість бронювань.

Також можна дослідити, що сприяло зростанню бронювань за роками (нові маркетингові кампанії, покращення сервісу чи інші чинники).

#### 6. Питання 4: Чи є відмінності в сезонності для різних каналів бронювання? 

```python
# Групування даних за датою та каналом бронювання і підрахунок бронювань
grouped_data = df.groupby(['date', 'distribution_channel']).size().unstack()

# Візуалізація
grouped_data.plot(kind='line', figsize=(12, 6), colormap='tab10')
plt.xlabel('Дата')
plt.ylabel('Кількість бронювань')
plt.title('Сезонність бронювань за каналами')
plt.legend(title='Канал бронювання')
plt.show();
```

**Висновки:** Канал TA/TO (Travel Agents / Tour Operators) є домінуючим за кількістю бронювань. Кількість бронювань коливається від 2000 до 5000 на місяць.
Відзначається чітка сезонність: пікові значення спостерігаються влітку (липень-серпень) та наприкінці року (грудень), а спад у зимові місяці (січень-лютий) і міжсезоння.
Corporate (Корпоративні бронювання) демонструє дуже низькі значення, менш ніж 500 бронювань на місяць. У нього відсутні чіткі сезонні коливання.
Канал Direct (Прямі бронювання) також має незначний обсяг бронювань (менше 500), але спостерігається деяке зростання протягом року.
У канала GDS (Global Distribution Systems) майже непомітний внесок у загальний обсяг бронювань.

Канал TA/TO показує значну залежність від сезонності. Туристичні оператори забезпечують сплеск активності в традиційно високі туристичні сезони (літо та зимові свята). Інші канали бронювання, як-от Corporate, Direct та GDS, не мають вираженої сезонності, їхній внесок у загальну картину є стабільно низьким.

Для збільшення кількості бронювань можна розглянути декілька ідей. Знижки та бонуси для туроператорів у високий сезон, або розробка спеціальних пакетів для зимових і літніх періодів. Розробка спеціальних корпоративних пропозиції, наприклад, знижки на групові бронювання або вигідні тарифи для регулярних клієнтів. Вихід на нові компанії через пряму рекламу, cold emailing або партнерство. Також організація конференцій чи тренінгів у партнерських готелях. Інвестування в покращення вебсайту чи мобільного додатку. Наприклад: швидкий та зручний процес бронювання, можливість бронювати акційні пропозиції, відгуки клієнтів для довіри. Запровадження системи знижок чи програма лояльності для клієнтів, які бронюють безпосередньо.

#### 7. Питання 5: Відсоток скасованих бронювань за сегментами ринку. 

```python
# Зведена таблиця для сегментів ринку
pivot_segment = pd.pivot_table(
    df, 
    values='is_canceled', 
    index='market_segment', 
    aggfunc='mean'
).reset_index()

# Додаємо колонку з відсотком скасувань
pivot_segment['cancellation_rate (%)'] = (pivot_segment['is_canceled'].astype(float) * 100)

# Видаляємо зайву колонку
pivot_segment = pivot_segment.drop(columns=['is_canceled'])

# Застосування градієнтного забарвлення
pivot_segment_styled = pivot_segment.style.background_gradient(subset=['cancellation_rate (%)'], cmap='Blues')

# Форматування таблиці з точністю до двох знаків після коми
pivot_segment_styled = pivot_segment_styled.format({'cancellation_rate (%)': '{:.2f}'})

# Вивід таблиці
pivot_segment_styled
```

**Висновки:** Найвищий відсоток скасувань мають групи — 64.23%. Це вказує на високу нестабільність у групових бронюваннях. Можливо, групові бронювання часто скасовуються через організаційні труднощі або непідтверджені плани. Другий за величиною рівень скасувань у онлайн-бронювань — 36.91%. #### 8. Питання 6: Залежність рівня скасувань бронювань по сегментах ринку від кількості днів між датою бронювання і датою прибуття (lead_time). 

```python
# Зведена таблиця: lead_time і рівень скасувань по сегментах ринку
pivot_segment_lead_time = df.groupby('market_segment').agg(
    avg_lead_time=('lead_time', 'mean'),
    cancellation_rate=('is_canceled', 'mean')
).reset_index()

# Додаємо колонку з відсотками скасувань
pivot_segment_lead_time['cancellation_rate (%)'] = pivot_segment_lead_time['cancellation_rate'] * 100

# Видаляємо зайву колонку з часткою
pivot_segment_lead_time = pivot_segment_lead_time.drop(columns=['cancellation_rate'])

# Округлення всіх числових значень до 2 знаків
pivot_segment_lead_time = pivot_segment_lead_time.round(2)

pivot_segment_lead_time
```

**Висновки:** Сегменти з тривалим часом бронювання, такі як Groups і Offline TA/TO, мають вищий рівень скасувань. Це може бути пов'язано з тим, що бронювання здійснюються заздалегідь, але з часом плани клієнтів змінюються. Сегменти Aviation та Complementary з коротким lead_time мають відносно низький або середній рівень скасувань, оскільки бронювання здійснюються ближче до дати поїздки, що зменшує ризик зміни планів.

Впровадити суворіші умови скасування для сегментів Groups, Offline TA/TO і Online TA. Також можна запропонувати додаткові знижки для бронювань із довгим lead_time за умови повної передоплати.

Можливо графік значно полегшить аналіз, наочно показуючи співвідношення між середнім часом бронювання та рівнем скасувань для кожного сегмента ринку. Він може швидко ідентифікувати сегменти з найбільшими проблемами чи аномаліями.

```python
# Візуалізація середнього lead_time та рівня скасувань
plt.figure(figsize=(10, 6))

# Створення двох осей для порівняння
ax1 = sns.barplot(data=pivot_segment_lead_time, x='market_segment', y='avg_lead_time', color='skyblue', label='Середній Lead Time')
ax1.set_ylabel('Середній Lead Time (дні)')
ax1.set_xlabel('Сегмент ринку')
plt.xticks(rotation=45)

# Додатковий графік для рівня скасувань
ax2 = ax1.twinx()
sns.lineplot(data=pivot_segment_lead_time, x='market_segment', y='cancellation_rate (%)', color='red', marker='o', label='Рівень скасувань (%)', ax=ax2)
ax2.set_ylabel('Рівень скасувань (%)')

# Заголовок і легенда
plt.title('Залежність рівня скасувань від Lead Time по сегментах ринку')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show();
```

**Висновки:** Чим довший Lead Time, тим вищий рівень скасувань. Це чітко видно на прикладі сегментів Groups, Offline TA/TO та Online TA. Короткий Lead Time (наприклад, у Aviation) знижує ймовірність скасування, хоча винятки можливі через специфіку сегмента.

Найпроблемніші сегменти: Groups - найдовший Lead Time та найвищий рівень скасувань, Online TA - гнучкі умови скасування роблять цей сегмент вразливим.

Найстабільніші сегменти: Direct - низький рівень скасувань і середній Lead Time, Complementary - низький рівень скасувань та короткий Lead Time.

Для покращення ситуації можна заохочувати клієнтів до підтвердження планів (наприклад, через бонуси чи знижки за раннє бронювання з передоплатою). Встановити певні обмеження для безкоштовного скасування, особливо для Online TA.
```python
# Додавання стовпця із загальною кількістю гостей
confirmed_bookings['total_guests'] = (
    confirmed_bookings['adults'] + 
    confirmed_bookings['children'] + 
    confirmed_bookings['babies']
)

# Створення зведеної таблиці
meal_of_hotel = confirmed_bookings.pivot_table(
    values=['total_guests'],
    index='meal',
    columns='hotel',
    aggfunc={
        'total_guests': 'mean'
    }
).round(2)

# Додавання градієнтного забарвлення до таблиці
styled_meal_of_hotel = meal_of_hotel.style.background_gradient(
    cmap="Reds",
    axis=None
)

# Виведення таблиці
styled_meal_of_hotel
```

А у курортних готелях значення трохи вищі, ніж у міських готелях, особливо для FB (2.24) та HB (2.10). Це свідчить про більшу популярність курортних готелів серед сімей або великих груп. Undefined (1.96) також має місце, але його вплив незначний. А ось у міських готелях немає клієнтів з невизначеним типом харчування.

#### 11. Питання 9: Аналіз співпадіння кімнати бронювання і кімнати призначення

```python
# Додавання стовпця з інформацією, чи тип заброньованої кімнати збігається з призначеною
df['room_mismatch'] = df['reserved_room_type'] != df['assigned_room_type']

# Частота призначення типів кімнат щодо заброньованих
pivot_room_correlation = df.pivot_table(
    values='adr',
    index='reserved_room_type',
    columns='assigned_room_type',
    aggfunc='count',
    fill_value=0
)

# Виведення результатів
pivot_room_correlation
```

**Висновки:** Номери типу A мають найбільший попит (34,408 бронювань). Потрібно забезпечити достатній запас номерів цього типу, оскільки їх перепризначення іноді веде до перепризначень інших типів (наприклад, D). Тип D часто призначається іншими типами. Це свідчить про те, що кількість номерів цього типу може бути недостатньою для задоволення попиту. Типи C, F, і G показують хорошу відповідність між заброньованими та призначеними номерами, що свідчить про оптимальне управління цими типами номерів.

Провести детальний аналіз попиту на типи D та A, оскільки ці типи часто перепризначаються. Можливо збільшити запас номерів типу D або адаптувати стратегію бронювання. Можна оптимізувати систему бронювань, щоб зменшити частоту перепризначень (наприклад, запровадити обмеження на перепризначення для популярних типів номерів). Також можна використовувати перепризначення як інструмент для апгрейду, пропонуючи клієнтам покращений тип номера (наприклад, з A на B або C) для підвищення лояльності клієнтів та доходів. Також, як варіант можна створити спеціальні пакети для типів номерів із низькою частотою бронювання (наприклад, E, F, H), щоб збільшити їхню популярність.

Можливо варто проаналізувати вплив перепризначень на середню денну ставку (ADR) що дозволить зрозуміти, чи перепризначення пов’язане зі зниженням або підвищенням доходів.

```python
# Вплив різниці між типами номерів на ADR
pivot_adr_mismatch = df.pivot_table(
    values='adr',
    index='room_mismatch',
    aggfunc='mean'
).round(2)

pivot_adr_mismatch
```

**Висновки:** Середня денна ставка вища, коли заброньований номер відповідає призначеному. Це свідчить, що клієнти, які отримують те, що замовили, платять більше, або це може бути пов’язано з тим, що такі бронювання частіше включають преміальні типи номерів. А ось для перепризначених номерів середня денна ставка значно нижча. Це може свідчити про зниження доходу через: надання клієнтам номерів нижчого класу. Або знижки, або компенсації, які надаються клієнтам у разі розбіжності між заброньованим і призначеним номером.

Необхідно проаналізувати причини перепризначень (наприклад, недостатня кількість певних типів номерів, надлишкові резерви номерів іншого типу). Можливо зменшення кількості перепризначень допоможе уникнути втрат доходу.


#### 12. Питання 10: У які дні тижня більш популярні бронювання номерів

```python
# Агрегуємо дані
total_weekend_nights = df['stays_in_weekend_nights'].sum()
total_week_nights = df['stays_in_week_nights'].sum()

# Дані для кругової діаграми
labels = ['Вихідні дні', 'Робочі дні']
sizes = [total_weekend_nights, total_week_nights]
colors = ['#66c2a5', '#fc8d62']

# Побудова кругової діаграми
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Розподіл ночей у вихідні та робочі дні (усі готелі)')
plt.show();
```

**Висновки:** Робочі дні займають більшу частину — 72.2%. Це свідчить про те, що більшість ночей бронюється саме на будні дні, що може вказувати на переважання бізнес-мандрівок або робочих поїздок. А вихідні ночі можуть бути популярні серед туристів, які подорожують для відпочинку, але їхня частка є меншою порівняно з робочими днями.

Можливо, якщо розглядаються готелі в ділових районах або великих містах, багато бронювань роблять люди, які приїжджають у справах. Вони зазвичай залишаються в готелях у робочі дні. Також багато готелів мають вищі ціни на вихідні, що може впливати на зменшення попиту у ці дні.

Варто розробити акційні пропозиції або спеціальні пакети для туристів, щоб підвищити завантаженість у вихідні.Також готелі можуть оптимізувати свої пропозиції для бізнес-аудиторії (наприклад, конференц-зали, послуги для бізнес-клієнтів). Можливо для більш детального аналізу варто дослідити розподіл ночей між міськими і курортними готелями.

```python
# Агрегуємо дані за типом готелю
grouped = df.groupby('hotel')[['stays_in_weekend_nights', 'stays_in_week_nights']].sum()

# Кількість типів готелів
num_hotels = len(grouped.index)

# Створюємо сітку під графіки (1 рядок, стільки колонок, скільки типів готелів)
fig, axes = plt.subplots(1, num_hotels, figsize=(6 * num_hotels, 6))

# Побудова кругових діаграм для кожного типу готелю
for ax, hotel_type in zip(axes, grouped.index):
    hotel_data = grouped.loc[hotel_type]
    
    labels = ['Вихідні дні', 'Робочі дні']
    sizes = [hotel_data['stays_in_weekend_nights'], hotel_data['stays_in_week_nights']]
    colors = ['#66c2a5', '#fc8d62']
    
    ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
    ax.set_title(f'Розподіл ночей для {hotel_type}')

# Корекція відступів
plt.tight_layout()
plt.show();
```

**Висновки:** Робочі дні займають більшу частину — 72.3%. Це свідчить про те, що міські готелі переважно обслуговують клієнтів у робочі дні, ймовірно, бізнес-клієнтів або людей, які подорожують у справах. А менший попит у вихідні може бути пов’язаний з тим, що туристи рідше обирають міські готелі для відпочинку. Частка робочих днів у курортних готелів трохи менша, ніж у міських готелях. Курортні готелі також популярні серед гостей у робочі дні, але, ймовірно, вони поєднують бізнес і відпочинок. А ось у вихідні дні частка у курортних готулів навпаки більше, ніж у міських. Це логічно, оскільки курортні готелі зазвичай орієнтовані на відпочинок, тому частка гостей у вихідні трохи вища. У обох готелів робочі дні домінують у розподілі ночей. Це свідчить про те, що більшість гостей користуються готелями в будні, незалежно від їх типу.

Міським готелям слід акцентувати увагу на послугах для бізнес-гостей: конференц-зали, бізнес-центри, швидка реєстрація. Для збільшення попиту у вихідні можна впровадити акції чи знижки для туристів. А для курортних готелів можна збільшити привабливість у будні через пакети послуг, що включають відпочинок і роботу (workation). Для вихідних корисно розробляти спеціальні пропозиції для сімейного відпочинку або коротких туристичних поїздок.

#### 13. Питання 11: Взаємозв’язок між кількістю ночей та середньою денного ставкою

```python
# Фільтрація тільки підтверджених бронювань
df = df[df['is_canceled'] == 0]

# Додавання колонки із загальною кількістю ночей
df['total_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']

# Налаштування графіка за допомогою Seaborn
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df,
    x='total_nights',
    y='adr',
    s=100,
    hue="adr",
    alpha=0.8
)

# Додавання підписів
plt.title('Взаємозв’язок між кількістю ночей та середньою денною ставкою (ADR)')
plt.xlabel('Загальна кількість ночей')
plt.ylabel('Середня денна ставка (ADR)')

# Показ графіка
plt.show();
```

**Висновки:** Чим довше перебування, тим нижчий середній ADR. Це може бути частиною стратегії готелю, щоб залучити клієнтів для довготривалого проживання. Для коротких бронювань ADR часто вищий, що вказує на попит на короткі, але комфортні проживання. Є кілька точок, де ADR дорівнює 0, незалежно від кількості ночей. Це може бути пов’язано із: помилками в даних або наявністю безкоштовних бронювань (наприклад, бонусні програми, корпоративні пакети). Потрібно перевірити дані для таких записів і з’ясувати, чи це помилки, чи спеціальні випадки.

Можна активніше пропонувати знижки для клієнтів із тривалими бронюваннями, щоб зберігати високу заповнюваність. Сегмент клієнтів із короткими бронюваннями, готових платити високі ADR, варто таргетувати преміум-пропозиціями.

Для залучення більшої кількості клієнтів можна запровадити різні рівні цін залежно від тривалості проживання. Наприклад вищі ціни для коротких бронювань із додатковими послугами, або знижки для тривалих проживань для стимулювання довготривалих бронювань. Також можна активно рекламувати преміальні номери для клієнтів, які бронюють на 1–3 ночі. Або можна розробити спеціальні пакети для клієнтів, які бронюють понад 10 ночей, наприклад, знижки на харчування чи додаткові послуги.
